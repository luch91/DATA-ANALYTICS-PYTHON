Dealing with Empty Cells

Option 1: Remove the rows with the empty data

```python
df = pd.read_csv("Black_friday_sales1.csv", header = 0)
df.describe()
df_new = df.dropna()
```

```python
df = pd.read_csv("Black_friday_sales1.csv", header = 0)
df_new = df.dropna()
df_new.describe()
```

Option 2: Replace Nan with 0

```python
df = pd.read_csv("Black_friday_sales1.csv", header = 0)
df_replace = df.fillna(0)
df_replace.describe()
```

Option 3: If the there are no values available fpr the empty cells, replace the NaN values with the mean, median or mode of all the values in the column.

```python
# replace with mean
df = pd.read_csv("Black_friday_sales1.csv", header = 0)
mean = df["Product_Category_1"].mean()
df_new = df.fillna(mean)
df_new.head()
```

```python
# replace with median
df = pd.read_csv("Black_friday_sales1.csv", header = 0)
median = df["Product_Category_1"].median()
df_new = df.fillna(median)
```

Dealing with Wrong Data

```python
import numpy as np
```

```python
# replace NaN with zero

df["Product_Category_3"].replace(np.nan, 0, inplace= True)
df["Product_Category_2"].replace(np.nan, 0, inplace= True)
df["Product_Category_1"].replace(np.nan, 0, inplace= True)
```

```python
# Replace wrong datatype(numerical) with integer

df["Stay_In_Current_City_Years"].replace("4+","4", inplace= True)
df.head()
```

```python
# replace intervals with integers

df["Age"].unique() #finds all the unique intervals
```

```python
df["Age"].replace(['46-50', '26-35', '36-45', '18-25', '51-55', '55+', '0-17'], ['50', '35', '45', '25', '55', '60', '17'], inplace= True)
df.head(10)
```

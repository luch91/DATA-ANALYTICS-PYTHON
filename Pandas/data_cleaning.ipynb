Indexing with Series

```python
# Using .loc[] for index-labeling

data = pd.Series([10,34,89,76,50], index=['a','b','c','d','e'])
print(data.loc['c'])
print(data.loc[['a','d']])  # access multiple elements by label
print(data.loc['a':'c'])  # access a range of elements by label
```

```python
# Using .iloc[] for integer-based indexing:

data = pd.Series([23,20,30,67,89,90,67])
print(data.iloc[2])  # access a specific element by integer position
print(data.iloc[[0,2]])  # Access multiple element by integer position
print(data.iloc[2:4])   # Access a range of elements by integer position
```

```python
# Boolean indexing

data = pd.Series([10,34,89,76,50], index=['a','b','c','d','e'])
print(data[data > 20])  # perform based on a condition
```

Slicing Series

the process of selecting a subset of rows or cloumns of a DataFrame or Series based on their position
usin the [] operator or .iloc[] method

```python
data = pd.Series([23,20,30,67,89,90,67])
sliced_data = data[1:5]
print(sliced_data)
```

```python
# using .iloc[]

data = pd.Series([23,20,30,67,89,90,67])
sliced_data = data.iloc[1:5]
sliced_data2 = data.iloc[1:]
sliced_data3 = data.iloc[:4]
print(sliced_data)
print(sliced_data2)
print(sliced_data3)

```
